# Setting Up Spark Session

A `SparkSession` is your entry point to working wit Spark and it comes witn wide range of configuration options. These options help you control resources, optimize data processing and customize logging for better debugging. 

## 1. Basic Spark Session Setup

Start by creating a `SparkSession` using `SparkSession.builder` API:


````python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
  .appName('Hello from Spark')\
  .getOrCreate()

print("Spark Session is ready 🚀")
````

This basic setup includes an `appName` which is usefule for indentifying Sparks distributed processes:

## 2. Configuration Options for `SparkSession`

**Memory and Resource Management**:

These Settings let you control memory allocation and CPU usage for Sparks distributed processes:

1. Executor Memory(`spark.executor.memory`): Sets the memory allocation for each executor.

2. Driver Memory(`spark.driver.memory`): Allocates memory for the driver.

3. Core Allocation(`spark.executor.cores`): Specifies the number of CPU cores per executor.

Read more about the configurations properties [here](https://spark.apache.org/docs/latest/configuration.html#application-properties)


Example Configuration:
```
spark = SparkSession.builder \
    .appName("Optimized App") \
    .config("spark.executor.memory", "2g") \
    .config("spark.driver.memory", "1g") \
    .config("spark.executor.cores", "2") \
    .getOrCreate()

```

**Shuffle and Storage Settings**:

1. Shuffle Partitions(`spark.sql.shuffle.partitions`): Configures the number of partitions for shuffle operations , often useful in joins aggregations . Learn more [here](https://spark.apache.org/docs/latest/sql-performance-tuning.html#other-configuration-options)

2. Storage level(`spark.storage.level`): Controls caching enabling more efficient reuse of DataFrames .see [spark.storage.level](https://spark.apache.org/docs/latest/rdd-programming-guide.html#persistence)

Example:

```
spark = SparkSession.builder \
    .appName("Shuffle Optimization") \
    .config("spark.sql.shuffle.partitions", "100") \
    .getOrCreate()
```

**Logging and Debugging**:

These Options enhance Sparks logging and debugging capabilities:

1. Log level(`spark.eventLog.enabled`):Enables event logging for monitoring jobs. Details on logging options are available [here](https://spark.apache.org/docs/latest/monitoring.html#spark-events)

2. Checkpoint Directory(`spar.checkpoing.dir`): Specifies a directory for Spark to store checkpoint dat. see [spark.checkpoint.dir](https://spark.apache.org/docs/latest/streaming-programming-guide.html#checkpointing) for streaming jobs.


Example 

```
spark = SparkSession.builder \
    .appName("Logging App") \
    .config("spark.eventLog.enabled", "true") \
    .config("spark.eventLog.dir", "/path/to/logs") \
    .getOrCreate()


```

**JDBC Connections and Database Configurations**:

For application needing database connnectivity:

1. JDBC URL(`spark.sql.sources.jdbc.url`):Provides the JDBC URL to connect to an external database see. [spark.sql.sources.jdbc](https://spark.apache.org/docs/latest/sql-data-sources-jdbc.html)

2. Connection TimeOut(`spark.network.timeout`): Sets the network timeout for connections. Learn more about [spark.network.timeout](https://spark.apache.org/docs/latest/configuration.html#networking).

Example

```
spark = SparkSession.builder \
    .appName("Database App") \
    .config("spark.sql.sources.jdbc.url", "jdbc:postgresql://localhost:5432/mydb") \
    .config("spark.network.timeout", "120s") \
    .getOrCreate()


```

**Miscellaneouis Options**:

1. Broadcast Join Threshold(`spark.sql.autoBroadcastJoinThreshold`):Controls the data size threshold for broadcasting joins .Learn more from [here](https://spark.apache.org/docs/latest/sql-performance-tuning.html#broadcast-hint-for-sql-queries)

Example

```
spark = SparkSession.builder \
    .appName("Misc Configurations") \
    .config("spark.local.dir", "/path/to/temp") \
    .config("spark.sql.autoBroadcastJoinThreshold", "10MB") \
    .getOrCreate()

```

## Step 3: Finalizing Your SparkSession Setup
Once configured, call .getOrCreate() to initialize the Spark session with your specified settings.


Each configuration option allows you to adjust Spark’s behavior to suit specific resource needs, logging preferences, and file storage formats, making SparkSession highly adaptable for various big data scenarios.
