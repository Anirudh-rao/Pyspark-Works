## 2. Meet Your New Best Friend: SparkSession
Every great PySpark adventure starts with a **SparkSession**. Think of it as your trusty sidekick, like Robin to Batman. It’s the entry point to everything Spark!



````python
from pyspark.sql import SparkSession

# Intialize the SparkSession 
spark = SparkSession.builder\
  .appName('Pyspark 101')\
  .getOrCreate()

print("Sparks in the House 🔥")
````

## 3. DataFrames: Like pandas, but on steroids 💪

If you’ve used pandas, you know DataFrames are your bread and butter for data manipulation. 
Well, guess what? PySpark has DataFrames too, but they can handle BIGGER data, and they do it with style. 😎

Let’s load a sample CSV file:

````python
import pandas as pd

# Read the CSV file using pandas with tab delimiter
pandas_df = pd.read_csv('/Workspace/Users/anirudh.work.13.12@outlook.com/Pyspark-Works/Data/marketing_campaign.csv', sep='\t')

# Convert pandas DataFrame to Spark DataFrame
df = spark.createDataFrame(pandas_df)
df.show(5)
````

## 4. DataFrames are Lazy....But in a Good Way:

Pysparks DataFrames are like a friend who promises to do the dishers but waits until the very last second. They are lazy which means they done actually do any work until you tell them to . This is called Lazy evaluation.


````python
df_filtered = df.filter(df['Marital_Status'] == 'Single')
````

Nothing happens .But when you force it to act ,by using action like `.show()` or `.collect()` , thats when it rolls up its sleeves and does the work.

````python
df_filtered.show()
````

## 5. RDD's : The Ancestors of DataFrames

Before Pyspark DataFrame , there were RDD's (Resilient Distributed Datasets). Imagnine RDDs as wise old wizard who knows all about data distribution but are as easy to work with as DataFrames.

While DataFrames are modern and optimzed for SQL-like Operations RDD's are bit more ...Hands-on , unless you want to fell you are casting spells on your data.


## 6. Transformations Vs Actions

1. **Transformations** : These are like making to-do list . You say 'Spark I want to select column, filter, groupby this data' .Spark nods but does not move just yet. These are **lazy Operations**.

  Examples: `. filter()`,`.select()`,`.groupBy()`.

2. **Actions** : This is you shouting, "DO THE THING" and Spark like well trained dog jumps into action. These are the operations that trigger Spark to actually execute transformations.

Example: `.show()`, `.count()`,`.collect()`.
