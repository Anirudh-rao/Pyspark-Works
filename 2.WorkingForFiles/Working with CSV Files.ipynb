# Working with CSV Files

When it comes to handling CSV Files, Pysark offers a range of felxible options. Whether you are reading or writing to a CSV , Pyspark's built in capabilites makes it easy to configure these operations exactly how you want.

## Practice Question

### Step 1: Start with SparkSession

As always start by setting up your `SparkSession`. This is your connections to Pysparks prowerfule data processing engine.

````python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
    .appName("CSV Hanlding in Pyspark")\
    .getOrCreate()
print("Spark is Ready!")
````

### Step 2: Reading a CSV File:

To read a CSV File use `spark.read.csv()` . Pyspark offers various options to customize how the CSV is read, so you can handle headers, delimiters,schemas and more

#### Basic CSV Reading
Here is a simple examples for loading a CSV with Default Setting:


````python
df = spark.read.csv('/Data/Wine.csv')
df.head(10)
````

**Common CSV Reading Options**:
Pyspark allows you to customize your csv read operations with these commonly used options:

1. **Header**: if your CSV has a header row, set `header=True` to use it as column names.
2. **InferSchema**: To automatically the data types of each columns set, `inferSchema = True`.
3. **Delimeter**: To use Customer delimiter(sucha as `;`or `|`) set the `sep` options.

Examples of reading CSV with options:

````python
df = spark.read.option("header", True)\
        .option("inferschema",True)\
        .option("sep",";")\
        .csv('/Data/Wine.csv')
print(df.head(10))
````

**Additional Reading Options**

* **NUll Values**: Use `nullvalues` to specify a string to interpret as `null` such as `"N\A"` or `NUll`.
* **Date Format**: Set `dateFormat` to speficy the date fromat for data fields.
* **Mode** : Determine how spark handles corrupt records. Options Include:
    * `PERMISSIVE(`default`): Puts corrupt records in a seperate column.
    * `DROPMALFORMED`: Drops rows with the bad records.
    * `FAILFAST` : Throws an error for corrupt records.

Examples with null values, date formats and mode options:

````python
# Reading CSV with more options
df = spark.read.option("header", True) \
               .option("inferSchema", True) \
               .option("nullValue", "N/A") \
               .option("dateFormat", "MM/dd/yyyy") \
               .option("mode", "DROPMALFORMED") \
               .csv('/Data/Wine.csv')

df.show(5)
````

## Writing a DataFrame to a CSV File:

Writing to a CSV is equally straightfoward with `df.write.csv()` , and Pyspark gives you several options for customizing how output is saved:

### Basic CSV Writing:
Here's the simplest way to save a DataFrame as a CSV:

```

# Basic CSV write
df.write.csv("/path/to/output_data")


```

**Common CSV Writing Options**
* **Header**: Set header=True to include column names in the first row.
* **Mode**: Controls the behavior if the output file or directory already exists.
* **overwrite**: Replaces existing data.
* **append**: Adds data to the existing file.
* **ignore**: Skips writing if the file exists.
* **error or errorifexists**: Throws an error if the file exists.

Example of writing a CSV with header and mode options:

```

# Writing CSV with options
df.write.option("header", True) \
        .mode("overwrite") \
        .csv("/path/to/output_data")



```

**Additional Writing Options**

* **Delimiter**: Set sep to specify a custom delimiter.
* **Compression**: Choose a compression method such as `gzip`, `bzip2`, `lz4`, `snappy`, or `deflate`.

Example with custom delimiter and compression:

````python
# Writing CSV with delimiter and compression
df.write.option("header", True) \
        .option("sep", ";") \
        .option("compression", "gzip") \
        .csv("/Workspace/Users/anirudh.work.13.12@outlook.com/Pyspark-Works/Data/Wine_Compressed.csv")
````
